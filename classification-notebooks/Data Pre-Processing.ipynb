```python
# Copy out original feature set for use with reading classification results later.
#educationFeatureSubset = df[['Education_Attainment', 'Age', 'Opinions_on_Censorship', 'Years_on_Internet', 'Web_Ordering', 'Marital_Status', 'Gender', 'Major_Geographical_Location', 'Race', 'Registered_to_Vote', 'Sexual_Preference']].copy()
# Copy out original feature set for use with reading classification results later.
#occupationFeatureSubset = df[['Major_Occupation','Age', 'Opinions_on_Censorship', 'Years_on_Internet', 'Web_Ordering', 'Marital_Status', 'Gender', 'Major_Geographical_Location', 'Race', 'Registered_to_Vote', 'Sexual_Preference']].copy()

educationFeatureSubset = df[['Education_Attainment', 'Age', 'Opinions_on_Censorship', 'Years_on_Internet', 'Web_Ordering', 'Marital_Status', 'Gender', 'Community_Building', 'Country']].copy()
occupationFeatureSubset = df[['Major_Occupation','Age', 'Opinions_on_Censorship', 'Years_on_Internet', 'Web_Ordering', 'Marital_Status', 'Gender', 'Community_Building', 'Country']].copy()

# Assign the index so that it matches that of the original df
educationFeatureSubset.set_axis(df.index, axis='index', inplace=True)
occupationFeatureSubset.set_axis(df.index, axis='index', inplace=True)
```

## Data Preparation for use in Classification


### Data Encoding
This section will prepare data by use of one hot encoding to convert all column values to numeric for use with classification algorithms.

```python
# Import LabelEncoder
from sklearn import preprocessing
#creating labelEncoder
le = preprocessing.LabelEncoder()

#creating labelEncoder
le = preprocessing.LabelEncoder()

encodedOccupationFeatureSubset = occupationFeatureSubset[occupationFeatureSubset.columns.difference(['Major_Occupation'])].apply(le.fit_transform)
encodedOccupationFeatureSubset['Major_Occupation'] = pd.Series(occupationFeatureSubset['Major_Occupation'], index=encodedOccupationFeatureSubset.index)
encodedOccupationFeatureSubset.head(5)

encodedEducationFeatureSubset = educationFeatureSubset[educationFeatureSubset.columns.difference(['Education_Attainment'])].apply(le.fit_transform)
encodedEducationFeatureSubset['Education_Attainment'] = pd.Series(educationFeatureSubset['Education_Attainment'], index=encodedEducationFeatureSubset.index)
encodedEducationFeatureSubset.head(5)
```

We now have our data prepared both through selecting an interesting feature set, ensuring values are transformed to numeric values for use with the three classification algorithms and also prearing the new dataframe for analysis. 

```python
#EXPORT to CSV the original feature subset for use with reading classification results later in other algorithms.
encodedEducationFeatureSubset.to_csv('educationFeatureSubset.csv', encoding='utf-8')
encodedOccupationFeatureSubset.to_csv('occupationFeatureSubset.csv', encoding='utf-8')
```
